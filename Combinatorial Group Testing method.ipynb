```python
import pandas as pd
import math
import numpy as np
import random

num_specimens = input('Enter the number of Specimens that you would like to test')
num_specimens = int(num_specimens)

q = int(math.log(num_specimens, 3))

# formula to calculate number of testing pools
num_pools = int((pow(q, 2) + 5*q)/2)

(row, col) = (num_pools, num_specimens)
temp_data = [col * [0]] * row

matrix_m = pd.DataFrame(temp_data)
matrix_m
```

```python
def numberToBase(n, b): # Credit to https://stackoverflow.com/questions/2267362/how-to-convert-an-integer-to-a-string-in-any-base
    if n == 0:
        return [0]
    digits = []
    while n:
        digits.append(int(n % b))
        n //= b
    return digits[::-1]
```

```python
# Converts base 10 numbers to radix 3
new_index = []
for i in range(0, len(matrix_m.loc[0])):
    new_index.append(list(reversed(numberToBase(i, 3))))

for index in new_index:
    while(len(index) != q):
        index.append(0)

new_index
```

```python
# split matrix into two matrices called B and C

# formula to claculate number of rows/tests for matrix B
num_pools = 3 * q

(row, col) = (num_pools, num_specimens)
temp_data = [col * [0]] * row

matrix_b = pd.DataFrame(temp_data)

label_row = {}
position = 0
value = 0

# this creates a dictionary so every row can has a <p,v> value
for i in range(0, len(matrix_b.index)):
    label_row[i] = {position: value}
    value += 1
    if(value == 3):
        position += 1
        value = 0

# this loop determines which specimens go to which pool/test
position = 0
count = 0
for row in range(0, len(matrix_b.index)):
    for col in range(0, len(matrix_b.loc[row])):
        if(label_row[row][position] == new_index[col][position]):
            matrix_b[col][row] = 1
    count += 1
    if(count%3 == 0):
        position += 1
        count = 0

matrix_b
```

```python
# formula to calculate number of row/test for matrix C

# q choose 2 formula 
num_pools = int(math.factorial(q)/(2 * math.factorial(q - 2)))

(row, col) = (num_pools, num_specimens)
temp_data = [col * [0]] * row

matrix_c = pd.DataFrame(temp_data)

label_row = {}
position = 0
value = 1

# this creates a dictionary so every row can has a <p,p'> value where p < p'
for i in range(0, len(matrix_c.index)):
    label_row[i] = {position: value}
    if(value == q - 1):
        position += 1
        value = position
    value += 1

# this loop determines which specimens go to which pool/test
position = 0
count = 0
for row in range(0, len(matrix_c.index)):
    arr = list(label_row[row].keys())
    position = arr[0]
    for col in range(0, len(matrix_c.loc[row])):
        if(new_index[col][label_row[row][position]] == new_index[col][position]):
            matrix_c[col][row] = 1
        
        
matrix_c
```

```python
# Use a random number generator to test code by randomly selecting defectives from the number of specimens
def select2Random():
    num1 = random.randrange(0, num_specimens - 1)
    num2 = random.randrange(0, num_specimens - 1)
    while(num1 == num2):
        num2 = random.randrange(0, num_specimens - 1)
    return [num1, num2]

def select1Random():
    num1 = random.randrange(0, num_specimens - 1)
    return [num1]

def selectRandom():
    return []

# Sampling random defectives that could either be 0, 1 or 2
def defective_tests():
    num = random.randrange(0, 2)
    if(num == 0):
        return selectRandom()
    elif(num == 1):
        return select1Random()
    else:
        return select2Random()

rand_defects = defective_tests()
```

```python
# calculates which rows contain the defective specimen/item
defective_test = []

for defect in rand_defects:
    temp = []
    for row in range(0, len(matrix_b[defect])):
        if(matrix_b[defect][row] == 1):
            temp.append(row)
    defective_test.append(temp)

defective_test

```
